```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
import ast 
import re
# 'Supported languages' 컬럼을 분석하여 각 언어의 빈도수를 계산합니다.
language_counter = Counter()

for languages in data['Supported languages']:
    # 문자열이 리스트 형태로 표현된 경우, ast.literal_eval을 사용해 리스트로 변환
    if isinstance(languages, str) and languages.startswith('[') and languages.endswith(']'):
        try:
            languages = ast.literal_eval(languages)
        except (ValueError, SyntaxError):
            # 문자열이 잘못된 리스트 형식인 경우, 일반 문자열로 취급
            pass
    
    # 언어 목록이 문자열인 경우
    if isinstance(languages, str):
        languages = [lang.strip() for lang in languages.split(',')]
    # 언어 목록이 리스트인 경우
    elif isinstance(languages, list):
        languages = [lang.strip() for lang in languages]
    
    language_counter.update(languages)

# 가장 많이 등장하는 언어들을 추출합니다.
most_common_languages = language_counter.most_common(10)

# 언어와 빈도수 리스트를 추출합니다.
languages, counts = zip(*most_common_languages)

# 데이터 시각화
plt.figure(figsize=(10, 8))
plt.barh(languages, counts)
plt.title('Frequency of Supported Languages')
plt.xlabel('Frequency')
plt.ylabel('Languages')
plt.show()
```

```python
# 언어 분리 및 카운트
language_counter = Counter()

# 각 게임의 지원 언어 리스트를 순회하며 언어의 빈도를 카운트
for language_list in data['Supported languages'].dropna():
    languages = [lang.strip().strip('[]') for lang in str(language_list).split(',')]
    language_counter.update(languages)

# 상위 10개 언어 추출
top_languages = [lang for lang, count in language_counter.most_common(10)]

# 언어별 판매량(리뷰 카운트) 데이터 준비
sales_by_language = {language: data[data['Supported languages'].str.contains(language, na=False)]['Review Count'].sum() for language in top_languages}

# 데이터프레임으로 변환
sales_df = pd.DataFrame(list(sales_by_language.items()), columns=['Language', 'Sales']).sort_values(by='Sales', ascending=False)

# 시각화
plt.figure(figsize=(12, 8))
sns.barplot(x='Sales', y='Language', data=sales_df, palette="Blues_d")
plt.title('Sales Volume by Top 10 Supported Languages')
plt.xlabel('Sales Volume')
plt.ylabel('Languages')
plt.show()
```
